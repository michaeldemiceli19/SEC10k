```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import time

header = {'User-Agent':'mdemiceli@gmail.com'}
```

```python
base_url = r'https://data.sec.gov/api/xbrl/companyfacts/CIK'
cik = '0001065280'

filing_url = base_url + cik + '.json'

content = requests.get(filing_url, headers=header)
decoded_content = content.json()
```

# UNDERSTANDING FILE SHAPE

```python
# these are the outer layers

for item in decoded_content['facts']:
    print(item)
    
print('-'*100)

for item in decoded_content['facts']['dei']:
    print(item)
    
print('-'*100)

 
for item in decoded_content['facts']['us-gaap']:
    print(item)
    
print('-'*100)
```

```python
# these are the inner layers

for item in decoded_content['facts']['us-gaap']['Revenues']:
    print(item)

print('-'*100)

for item in decoded_content['facts']['us-gaap']['Revenues']['units']:
    print(item)

print('-'*100)

for item in decoded_content['facts']['us-gaap']['Revenues']['units']['USD']:
    print(item)

print('-'*100)
```

# DATA CLEANING

```python
json_file = open(r'sec_' + cik + r'.json', 'w')
json.dump(decoded_content, json_file, indent=3)

with open(r'sec_' + cik + r'.json') as sec:
    data = json.load(sec)
```

```python
sec_data = {}
sec_data['headers'] = []
sec_data['categories'] = []
sec_data['data'] = []

for separator in data['facts']:
    
    for category in data['facts'][separator]:
        sec_data['categories'].append(category)
        
        for item in data['facts'][separator][category]:
            units = data['facts'][separator][category]['units']
            
            for vals in data['facts'][separator][category]['units']:
                    for subvals in data['facts'][separator][category]['units'][vals]:
                        end_date = subvals['end']
                        fp = subvals['fp']
                        fy = subvals['fy']
                        val = subvals['val']
                        
                        sec_data['headers'].append(str(f'{fy}{fp}'))
                        sec_data['data'].append(val)

            
        
sec_data
```
